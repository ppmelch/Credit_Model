```python
y = df["loan_status"]
```

```python
X = df[['age', 'occupation_status', 'years_employed', 'annual_income',
       'credit_score', 'credit_history_years', 'savings_assets',
       'current_debt', 'defaults_on_file', 'delinquencies_last_2yrs',
       'derogatory_marks', 'product_type', 'loan_intent', 'loan_amount', 'interest_rate',
       'debt_to_income_ratio', 'loan_to_income_ratio',
       'payment_to_income_ratio', 'loan_status']]

X = X.drop(columns=['interest_rate', 'loan_status'])

X = pd.get_dummies(X , drop_first= True)
```

```python
X_train , X_test , y_train , y_test =  train_test_split(
    X , y , test_size = 0.25 , random_state= 42)
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]  # 🔥 esto es PD

df['predicted_pd'] = model.predict_proba(X)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix

roc = roc_auc_score(y_test, y_pred_proba)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```

```python
print(roc)
print(acc)
print(cm)
```
